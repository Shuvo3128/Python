```python
squares = [x**2 for x in range(10)]
__builtins__.print(squares)
```

```python
evens = [x for x in range(10) if x % 2 == 0]
__builtins__.print(evens)
```

```python
data = 'i love data science'
[x for x in data]
```

```python
data = 'i love data science'
data = data.split()
[x for x in data if x != 'i']
```

```python
squares = [i**2 for i in range(1, 6)]
__builtins__.print(squares)
```

```python
numbers = [1, 2, 3, 4, 5, 6]
evens = [n for n in numbers if n % 2 == 0]
print(evens)
```

```python
names = ["ali", "sara", "john"]
upper_names = [name.upper() for name in names]
print(upper_names)
```

```python
nums = [1, 2, 3]
result = [n * 10 for n in nums]
print(result)
```

```python
words = ["python", "java", "c"]
lengths = [len(word) for word in words]
print(lengths)
```

```python
nums = [2, 5, 8, 1, 9]
big = [n for n in nums if n > 5]
print(big)
```

```python
letters = [ch for ch in "python"]
print(letters)
```

```python
nums = [1, 2, 3, 4, 5]
result = ["even" if n % 2 == 0 else "odd" for n in nums]
print(result)
```

```python
pairs = [(i, j) for i in range(1, 4) for j in range(1, 3)]
print(pairs)
```

```python
text = "python"
no_vowels = [ch for ch in text if ch not in "aeiou"]
print(no_vowels)
```
