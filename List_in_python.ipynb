```python
names = ["Ali", "Ahmed", "Sara", "John"]

for name in names:
    print(name)
```

```python
numbers = [10, 20, 30, 40]
total = 0

for n in numbers:
    total += n

print("Sum =", total)
```

```python
numbers = [5, 12, 3, 20, 7]
largest = numbers[0]

for n in numbers:
    if n > largest:
        largest = n

print("Largest =", largest)
```

```python
fruits = ["apple", "banana"]
fruits.append("mango")
fruits.append("orange")

print(fruits)
```

```python
numbers = [1, 4, 7, 8, 10, 13]
count = 0

for n in numbers:
    if n % 2 == 0:
        count += 1

print("Even numbers count =", count)
```
