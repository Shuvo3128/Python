```python
marks = (85, 90, 78)

for i in range(len(marks)):
    print("Index:", i, "Value:", marks[i])
```

```python
students = (
    ("Ali", 85),
    ("Sara", 90),
    ("John", 78)
)

for name, score in students:
    print(name, score)
```
