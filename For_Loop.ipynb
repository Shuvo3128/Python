```python
total = 0
for num in range(10, 1001):
  is_prime =True

  if num < 2:
    is_prime = False
  else:
    for i in range(2, num):
      if num % i == 0:
        is_prime = False
        break
  if is_prime:
    total += num
print(total)
```

```python
n = 7
factorial = 1

for i in range(1, n + 1):
    factorial *= i

print("Factorial of", n, "is", factorial)
```

```python
num = 27956240710
reverse_num = 0

while num > 0:
    digit = num % 10
    reverse_num = reverse_num * 10 + digit
    num //= 10

print("Reverse number:", reverse_num)
```
